```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(kredi.text,'lxml')
```

```python
print(s.prettify())
```

```python
carros = s.find_all('p')
```

```python
carros
```

```python
carro = carros[0]
```

```python
carro
```

```python
carro.get_text()
```

```python
lista_carros_bruto = [carro.get_text() for carro in carros]
```

```python
lista_carros_bruto
```

```python
copia_carros_bruto = lista_carros_bruto
modelos = []
transmision = []
precio = []
kilometros = []
metodo_pago = []
```

```python
while len(lista_carros_bruto) > 0:
    for x in range(5):
        if x == 0:
            modelos.append(copia_carros[x])
        elif x == 1:
            transmision.append(copia_carros[x])
        elif x == 2:
            precio.append(copia_carros[x])
        elif x == 3:
            kilometros.append(copia_carros[x])
        elif x == 4:
            metodo_pago.append(copia_carros[x])
    del copia_carros[0:5]
            
```

```python
for modelo_carro in modelos:
    print(modelo_carro)
```

```python
dict_carros = [dict() for i in range(len(modelos))]
```

```python
dict_carros
```

```python
for i in range(len(modelos)):
    dict_carros[i].update(Modelo= modelos[i])
    dict_carros[i].update(Transmision= transmision[i])
    dict_carros[i].update(Precio=precio[i])
    dict_carros[i].update(Kilometros=kilometros[i]) 
    dict_carros[i].update(Metodo=metodo_pago[i])
```
